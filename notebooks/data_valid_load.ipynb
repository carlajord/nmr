```python
import os, sys
import numpy as np, pandas as pd
import scipy.io
```

```python
DATA_PATH = os.path.join(os.path.dirname(""), "..", "data")
valid_data_path = os.path.join(DATA_PATH, "ValidationData_vol2")
```

```python
cases_idx = [19]#, 20, 22, 24, 28, 32]
data_len = len(cases_idx)
```

```python
mat_contents = {}
for idx, idx_data in enumerate(cases_idx):
    # load .mat file
    filename = os.path.join(valid_data_path, f"case0{idx_data}_segment00_result.mat")
    mat_contents[idx] = scipy.io.loadmat(filename)
```

```python
len_mixed = 19
```

Get all MQIs

```python
mqi_basis = []
mqi_mixed = []

for idx in range(data_len):
    for idx_mixed in range(len_mixed):
        mixed_data = mat_contents[idx]['mixed'][0][idx_mixed]
        mqi_mixed.append(mixed_data[2][0])
```

```python
mqi_mixed = np.concatenate(mqi_mixed).tolist()
```

```python
np.mean(mqi_mixed), np.max(mqi_mixed), np.min(mqi_mixed), np.median(mqi_mixed)
```
